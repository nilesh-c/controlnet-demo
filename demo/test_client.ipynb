```python
import requests
import io
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# API Endpoint
API_URL = "http://localhost:8000/generate"

image_path = "../test_imgs/mri_brain.jpg"

params = {
    "prompt": "mri brain scan",
    "a_prompt": "good quality",
    "n_prompt": "low quality, bad anatomy, unrealistic"
}

# Read Image and Send Request
with open(image_path, "rb") as image_file:
    files = {"file": image_file}
    response = requests.post(API_URL, files=files, data=params)

if response.status_code == 200:
    # Read and display the generated image
    image_bytes = io.BytesIO(response.content)
    image = Image.open(image_bytes)
    plt.imshow(image)
    plt.axis("off")
    plt.show()
else:
    print(f"Error: {response.status_code} - {response.text}")
```
